### Image Shape and batch size initialization

```python
img_shape = (254,254,3)
batch = 16
```

## Configuring ImageDataGenerator for loading the images efficiently

```python
train_image_gen = ImageDataGenerator(rotation_range=180,width_shift_range=0.1,
                                     height_shift_range=0.1,
                                     horizontal_flip=True,
                                     vertical_flip=True,
                                     rescale=1.0/255,
                                     zoom_range=0.1,
                                     fill_mode='nearest',
                                     validation_split=0.2)

test_image_gen = ImageDataGenerator(rescale=1.0/255)
```

```python
train_generator = train_image_gen.flow_from_directory(train_path, 
                                                      target_size=(254,254),
                                                      color_mode='rgb',
                                                      class_mode='binary',
                                                      batch_size=batch,
                                                      shuffle=True,
                                                      seed=19,
                                                      subset='training')

val_generator = train_image_gen.flow_from_directory(train_path, 
                                                      target_size=(254,254),
                                                      color_mode='rgb',
                                                      class_mode='binary',
                                                      batch_size=batch,
                                                      shuffle=True,
                                                      seed=19,
                                                      subset='validation')

test_generator = test_image_gen.flow_from_directory(test_path, 
                                                    target_size=(254,254), 
                                                    color_mode='rgb', 
                                                    class_mode='binary',
                                                    shuffle=False,
                                                    batch_size=batch)
```

```python
input_tensor = Input(shape=img_shape)
model = Xception(input_tensor=input_tensor, pooling='None')
```

```python
last_layer = model.get_layer('block14_sepconv2_act').output
x = Flatten(name='flatten')(last_layer)
# x = Dense(512, activation='relu', name='dense1')(x)
x = Dropout(0.4, name='dropout1')(x)
x = Dense(1024, activation='relu', name='dense2')(x)
# x = Dropout(0.4, name='dropout2')(x)
out = Dense(1, activation='sigmoid', name='output')(x)
model = Model(input_tensor, out)

for layer in model.layers:  # add extra 10 with your layers
    layer.trainable=True

# model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics='accuracy')
```

```python
# log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
# compare with 0.7052, 0.6144, 0.6306

early_stop = EarlyStopping(patience=10,verbose=1,monitor='val_accuracy',mode='min')
# tensorboard = TensorBoard(log_dir=log_dir, histogram_freq=1,write_graph=True,update_freq='epoch')

checkpoint = ModelCheckpoint('./{epoch:02d}_{val_accuracy:.04f}.h5', 
                             save_weights_only=False,
                             monitor='val_accuracy',
                             verbose=1,
                             save_best_only=True,
                             mode='max')

history = model.fit(train_generator, epochs=50, verbose=1,callbacks=[early_stop, checkpoint],
                  validation_data=test_generator)
```

# Evaluation of trained model.

```python
loaded_model = load_model('C://Users//kunal//Documents//Jupyter//CE888//models//test_acc_0.8988.h5')
```

```python
loaded_model.evaluate(test_generator, verbose=1)
pred = (loaded_model.predict(test_generator, verbose=1) > 0.5).astype("int32")
conf_mat = confusion_matrix(test_generator.classes, pred.ravel())
```

```python
sns.heatmap(conf_mat,annot=True,cmap='Blues',fmt='d')
print(classification_report(test_generator.classes, pred.ravel()))
```
